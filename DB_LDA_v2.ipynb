```python
Features=df.drop('attributeClass', axis=1)
Label=df['attributeClass']
```

```python
Features.shape
```

```python
Features_T=Features.T
#Features_T.columns= Features_T.iloc[0]
#Features_T.columns
```

```python
height, width = Features.shape
unique_classes = np.unique(Label)
unique_classes
```

```python
num_classes = len(unique_classes)
scatter_train = np.cov(Features_T)*(height - 1)
scatter_within = 0
```

```python
for i in range(num_classes):
 class_items = np.flatnonzero(Label == unique_classes[i])
 scatter_within = scatter_within + np.cov(Features_T[class_items]) * (len(class_items)-1)
 
scatter_between = scatter_train - scatter_within
```

```python
#Calculating Eigenvalues and Eigenvectors of the covariance matrix
eigen_values, eigen_vectors = np.linalg.eigh(np.linalg.pinv(scatter_within).dot(scatter_between))
#print(eig_vectors.shape)
#pc = Features.dot(eig_vectors[:,::-1][:,:self.n_components])
```

```python
#sort the eigenvalues in descending order
sorted_index = np.argsort(eigen_values)[::-1]
sorted_eigenvalue = eigen_values[sorted_index]
#similarly sort the eigenvectors
sorted_eigenvectors = eigen_vectors[:,sorted_index]
sorted_eigenvectors
```

```python
# select the first n eigenvectors, n is desired dimension
# of our final reduced data.
n_components = 1 #you can select any number of components.
eigenvector_subset = sorted_eigenvectors[:,0:n_components]
eigenvector_subset
```

```python
#Transform the data
Features_reduced = np.dot(eigenvector_subset.transpose(),Features.transpose()).transpose()
Features_reduced
```

```python
LDA_df = pd.DataFrame(Features_reduced)
LDA_df
```

#Split Data into Training/Testing

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(LDA_df, Label, test_size=0.3, random_state=1)
```

#Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred_train = clf.predict(X_train)
```

```python
print("Training-set accuracy (in %):",metrics.accuracy_score(y_train, y_pred_train)*100)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="entropy", max_depth=8)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

DT_Results = ['LDA', 'DT']
# Model Accuracy, how often is the classifier correct?
DT_Results.append(metrics.accuracy_score(y_test, y_pred)*100)
DT_Results.append(metrics.precision_score(y_test,y_pred,average='micro')*100)
DT_Results.append(metrics.f1_score(y_test,y_pred,average='micro')*100)
DT_Results.append(metrics.recall_score(y_test,y_pred,average='micro')*100)

print("Decision Tree model accuracy (in %):", metrics.accuracy_score(y_test, y_pred)*100)
print("Decision Tree model Precision Score (in %):", metrics.precision_score(y_test,y_pred,average='micro')*100)
print("Decision Tree model f1 Score (in %):", metrics.f1_score(y_test,y_pred,average='micro')*100)
print("Decision Tree model Recall Score (in %):", metrics.recall_score(y_test,y_pred,average='micro')*100)
```

```python
db_lda=pd.DataFrame(columns=['Method','Classifier','Accuracy','Precision_Score','F1-Score','Recall_Score'])
```

```python
db_lda.loc[len(db_lda)] = DT_Results
```
