```python
Features=ids_df.drop(['protocol_type','service','flag','Label'], axis=1)
Label=ids_df['Label']
```

# PCA

```python
# mean Centering the data  
Features_meaned = Features - np.mean(Features , axis = 0)
Features_meaned
```

```python
# Calculate the co-variance matrix of the mean-centered data.
cov_matrix = np.cov(Features_meaned, rowvar = False)
```

```python
#Calculating Eigenvalues and Eigenvectors of the covariance matrix
eigen_values , eigen_vectors = np.linalg.eigh(cov_matrix)
```

```python
#sort the eigenvalues in descending order
sorted_index = np.argsort(eigen_values)[::-1]
 
sorted_eigenvalue = eigen_values[sorted_index]
#similarly sort the eigenvectors 
sorted_eigenvectors = eigen_vectors[:,sorted_index]
sorted_eigenvectors
```

```python
# select the first n eigenvectors, n is desired dimension
# of our final reduced data.
 
n_components = 12 #you can select any number of components.
eigenvector_subset = sorted_eigenvectors[:,0:n_components]
eigenvector_subset
```

```python
#Transform the data 
Features_reduced = np.dot(eigenvector_subset.transpose(),Features_meaned.transpose()).transpose()
Features_reduced
```

```python
PCA_df = pd.DataFrame(Features_reduced)
PCA_df
```

```python
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics
```

# Split the dataset to Test/Train

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(PCA_df, Label, test_size=0.3, random_state=1)
```

# Decision Tree 

```python
# Import Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred_train = clf.predict(X_train)
```

```python
print("Decision Tree Model Accuracy with training data (in %):",metrics.accuracy_score(y_train, y_pred_train)*100)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="entropy", max_depth=8)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
DT_Results = ['PCA', 'DT']
# Model Accuracy, how often is the classifier correct?
DT_Results.append(metrics.accuracy_score(y_test, y_pred)*100)
DT_Results.append(metrics.precision_score(y_test,y_pred,average='micro')*100)
DT_Results.append(metrics.f1_score(y_test,y_pred,average='micro')*100)
DT_Results.append(metrics.recall_score(y_test,y_pred,average='micro')*100)

print("Decision Tree model accuracy (in %):", metrics.accuracy_score(y_test, y_pred)*100)
print("Decision Tree model Precision Score (in %):", metrics.precision_score(y_test,y_pred,average='micro')*100)
print("Decision Tree model f1 Score (in %):", metrics.f1_score(y_test,y_pred,average='micro')*100)
print("Decision Tree model Recall Score (in %):", metrics.recall_score(y_test,y_pred,average='micro')*100)
```

```python
ids_pca=pd.DataFrame(columns=['Method','Classifier','Accuracy','Precision_Score','F1-Score','Recall_Score'])
```

```python
ids_pca.loc[len(ids_pca)] = DT_Results
ids_pca
```
