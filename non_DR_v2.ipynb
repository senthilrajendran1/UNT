```python
Features=df.drop('NSP', axis=1)
Label=df['NSP']
```

# Split Data into Training/Testing

```python
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(Features, Label, test_size=0.3, random_state=1)
```

# Decision Tree

```python
# Import Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics 

# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred_train = clf.predict(X_train)
print("Decision Tree Model Accuracy with training data (in %):",metrics.accuracy_score(y_train, y_pred_train)*100)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="entropy", max_depth=8)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

DT_Results = ['non_dr', 'DT']
# Model Accuracy, how often is the classifier correct?
DT_Results.append(metrics.accuracy_score(y_test, y_pred)*100)
DT_Results.append(metrics.precision_score(y_test,y_pred,average='micro')*100)
DT_Results.append(metrics.f1_score(y_test,y_pred,average='micro')*100)
DT_Results.append(metrics.recall_score(y_test,y_pred,average='micro')*100)

print("Decision Tree model accuracy (in %):", metrics.accuracy_score(y_test, y_pred)*100)
print("Decision Tree model Precision Score (in %):", metrics.precision_score(y_test,y_pred,average='micro')*100)
print("Decision Tree model f1 Score (in %):", metrics.f1_score(y_test,y_pred,average='micro')*100)
print("Decision Tree model Recall Score (in %):", metrics.recall_score(y_test,y_pred,average='micro')*100)
```

```python
ctg_non_dr=pd.DataFrame(columns=['Method','Classifier','Accuracy','Precision_Score','F1-Score','Recall_Score'])
```

```python
ctg_non_dr.loc[len(ctg_non_dr)] = DT_Results
ctg_non_dr
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB() 
gnb.fit(X_train, y_train) 

y_pred_train = gnb.predict(X_train)

print('Training-set accuracy(in %):', metrics.accuracy_score(y_train, y_pred_train)*100)
```

```python
# making predictions on the testing set 
y_pred = gnb.predict(X_test)

# comparing actual response values (y_test) with predicted response values 

NB_Results = ['non_dr', 'NB']
# Model Accuracy, how often is the classifier correct?
NB_Results.append(metrics.accuracy_score(y_test, y_pred)*100)
NB_Results.append(metrics.precision_score(y_test,y_pred,average='micro')*100)
NB_Results.append(metrics.f1_score(y_test,y_pred,average='micro')*100)
NB_Results.append(metrics.recall_score(y_test,y_pred,average='micro')*100)

print("Naive Bayes model accuracy (in %):", metrics.accuracy_score(y_test, y_pred)*100)
print("Naive Bayes model Precision Score (in %):", metrics.precision_score(y_test,y_pred,average='micro')*100)
print("Naive Bayes model f1 Score (in %):", metrics.f1_score(y_test,y_pred,average='micro')*100)
print("Naive Bayes model Recall Score (in %):", metrics.recall_score(y_test,y_pred,average='micro')*100)
```

```python
ctg_non_dr.loc[len(ctg_non_dr)] = NB_Results
ctg_non_dr
```

# Random Forest

```python
ctg_non_dr.loc[len(ctg_non_dr)] = RF_Results
ctg_non_dr
```
